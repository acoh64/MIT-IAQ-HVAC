```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import datetime as dt
%matplotlib inline
```

# This files demonstrates how to extract and plot data from the data files

```python
HOMEDIR = os.getcwd()
DATA_DIR = "data"
room_name = "Classroom1"
filename_clean = os.path.join(HOMEDIR, DATA_DIR, room_name+"_clean.parquet")
filename_raw = os.path.join(HOMEDIR, DATA_DIR, room_name+"_raw.parquet")
```

```python
# Read data into a dataframe
data_clean = pd.read_parquet(filename_clean, engine='pyarrow')
data_raw = pd.read_parquet(filename_raw, engine='pyarrow')
print(data_raw.keys())
data_raw.head()
```

# Plot $\text{CO}_{2}$ concentration time series
To plot other time series, simply replace "CO₂ Concentration (ppm)" with the column of the data file that you want to plot (all column names are printed above)

```python
fig, axs = plt.subplots(2, 1)
fig.set_figwidth(10)
fig.set_figheight(6)
axs[0].plot(data_raw["CO₂ Concentration (ppm)"])
axs[0].set_xlabel("Date")
axs[0].set_ylabel("Raw CO$_{2}$ concentration (ppm)")
axs[1].plot(data_clean["CO₂ Concentration (ppm)"])
axs[1].set_xlabel("Date")
axs[1].set_ylabel("Cleaned CO$_{2}$ concentration (ppm)")

fig.autofmt_xdate()

fig.tight_layout()
plt.show()
```
